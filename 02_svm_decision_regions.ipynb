```python
import numpy as np
import matplotlib.pyplot as plt

# Define coefficients satisfying sum condition: β_1 + β_2 = 1
beta_0 = 0.2
beta_1 = 0.6
beta_2 = 0.4  # Ensuring β_1 + β_2 = 1

# Generate x1 values
x1 = np.linspace(-1.5, 1.5, 100)
x2 = (-beta_0 - beta_1 * x1) / beta_2  # Solve for x2 (Decision Boundary)

# Select three specific points
x1_on = -1.0
x2_on = (-beta_0 - beta_1 * x1_on) / beta_2  # On decision boundary

x1_above = -0.5
x2_above = x2_on + 0.2  # Slightly above the line

x1_below = 0.5
x2_below = x2_on - 3.2  # Slightly below the line

# Compute decision function values
f_on = beta_0 + beta_1 * x1_on + beta_2 * x2_on
f_above = beta_0 + beta_1 * x1_above + beta_2 * x2_above
f_below = beta_0 + beta_1 * x1_below + beta_2 * x2_below

# Print values
print(f"Point on decision boundary: ({x1_on:.2f}, {x2_on:.2f}),  f(x) = {f_on:.2f}")
print(f"Point above decision boundary: ({x1_above:.2f}, {x2_above:.2f}),  f(x) = {f_above:.2f} > 0")
print(f"Point below decision boundary: ({x1_below:.2f}, {x2_below:.2f}),  f(x) = {f_below:.2f} < 0")

# Create plot
plt.figure(figsize=(5, 5))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot decision boundary
plt.plot(x1, x2, 'k-', label=r'$\beta_0 + x^T \beta = 0$')

# Plot points
plt.scatter(x1_on, x2_on, color='orange', label=r'$ \beta_0 + x^T \beta = 0$', s=100)
plt.scatter(x1_above, x2_above, color='blue', label=r'$ \beta_0 + x^T \beta > 0$', s=100)
plt.scatter(x1_below, x2_below, color='red', label=r'$ \beta_0 + x^T \beta < 0$', s=100)

# Add equation annotation
plt.text(x1_above - 0.3, x2_above, r'$\beta_0 + x^T \beta > 0$', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.text(x1_below - 0.3, x2_below, r'$\beta_0 + x^T \beta < 0$', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Labels
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.legend()
plt.grid(True, linestyle=':', linewidth=0.5)

plt.show()
```
