```python
import numpy as np
import matplotlib.pyplot as plt

# Define coefficients satisfying sum condition: β_1 + β_2 = 1
beta_0 = 0.2
beta_1 = 0.6
beta_2 = 0.4  # Ensuring β_1 + β_2 = 1

# Generate x1 values
x1 = np.linspace(-1, 1, 100)
x2 = (-beta_0 - beta_1 * x1) / beta_2  # Solve for x2

# Generate sample points above and below the line
np.random.seed(42)
x1_below = np.random.uniform(-1, 1, 15)
x2_below = (-beta_0 - beta_1 * x1_below) / beta_2 - np.random.uniform(0.1, 0.3, 15)

x1_above = np.random.uniform(-1, 1, 15)
x2_above = (-beta_0 - beta_1 * x1_above) / beta_2 + np.random.uniform(0.1, 0.3, 15)

# Print equation dynamically
equation = f"{beta_0} + {beta_1}x₁ + {beta_2}x₂ = 0  with  β₁ + β₂ = 1"
print("Equation of the decision boundary:", equation)

# Create plot
plt.figure(figsize=(5, 5))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Plot decision boundary
plt.plot(x1, x2, 'k-', label=r'$\beta_0 + \beta_1 x_1 + \beta_2 x_2 = 0$')

# Plot points
plt.scatter(x1_below, x2_below, color='red', label='Below Line', alpha=0.7)
plt.scatter(x1_above, x2_above, color='blue', label='Above Line', alpha=0.7)


# Labels
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.legend()
plt.grid(True, linestyle=':', linewidth=0.5)

plt.show()
```
